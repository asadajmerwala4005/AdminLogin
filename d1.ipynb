Now that we have the full DataFrame, here's how you can filter it. Since there's no 'age' column, I'll demonstrate filtering for 'math score' greater than 70.

```python
filtered_df = df[df['math score'] > 70]
print(f"Number of students with math score > 70: {len(filtered_df)}")
display(filtered_df.head())
```

```python
students_with_high_math_score = df[df['math score'] > 40]
print(f"Number of students with math score > 40: {len(students_with_high_math_score)}")
display(students_with_high_math_score.head())
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame to show only female students, for example
female_students = df[df['gender'] == 'female']
print(f"Number of female students: {len(female_students)}")

# Or, to get counts of each gender
gender_counts = df['gender'].value_counts()
print("\nGender distribution:\n", gender_counts)

# Now, let's create a bar chart to visualize the gender distribution
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Display the math score column
print("Math Scores:\n", df['math score'].head())

# Get summary statistics for math score
print("\nMath Score Summary Statistics:\n", df['math score'].describe())

# Create a histogram to visualize the distribution of math scores
plt.figure(figsize=(8, 6))
df['math score'].hist(bins=10, edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
